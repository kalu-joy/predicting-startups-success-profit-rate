# Part 2 -- Exploratory Data Analysis (EDA)

```python
import import_ipynb
import preprocessing 
from preprocessing import startups_df
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

## Average Profit by State

```python
average_profit = startups_df.groupby('State')['Profit'].mean()

# Plot line plot
plt.plot(average_profit.index, average_profit.values, marker='D', color='green', linestyle='-')

# Add labels and title
plt.xlabel('State')
plt.ylabel('Average Profit')
plt.title('Average Profit by State')

# Show plot
plt.show()
```

## Average Marketing Spend by state

```python
average_marketing_spend = startups_df.groupby('State')['Marketing Spend'].mean()

#bar plot
plt.bar(average_marketing_spend.index, average_marketing_spend.values, color='green')
plt.xlabel('State')
plt.ylabel('Average Marketing Spend')
plt.title('Average Marketing Spend by State')
plt.show()
```

```python
average_RandD_spend = startups_df.groupby('State')['R&D Spend'].mean()

#bar plot
plt.bar(average_RandD_spend.index, average_RandD_spend.values, color='green')
plt.xlabel('State')
plt.ylabel('Average R&D Spend')
plt.title('Average R&D Spend by State')
plt.show()
```
